```python
def uefa_country_coeff(year):
    if type(year) != str:
        year = str(year)
        
    url = 'https://www.uefa.com/memberassociations/uefarankings/country/#/yr/' + year
    
    edgedriver = "Selenium/msedgedriver" # path to the edgedriver executable
    os.environ["webdriver.edge.driver"] = edgedriver
    
    driver = webdriver.Edge(edgedriver)
    driver.get(url)
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    table_soup = soup.find('div', id = 'calendar-container') # return whole table element
    
    driver.close()
    driver.quit()
    
    return year, table_soup    
```

```python
def save_soup(year, soup_obj):
    file_name = "uefa club coefficients/" + year + "_uefa_club_coeff.txt"
    
    with open(file_name, 'w', encoding = 'utf8') as f:
        f.write(str(soup_obj))
        f.close()
        
    print("Saved UEFA coefficient table for {}".format(year))     
```

```python
def loop_thru_years(first_year, last_year):
    year_list = list(range(first_year, last_year + 1))
    
    for i in year_list:
        try:
            year, soup_obj = uefa_country_coeff(i)
            save_soup(year, soup_obj)
        except:
            pass
```
