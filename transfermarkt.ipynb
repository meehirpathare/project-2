```python
def player(url):

    edgedriver = "Selenium/msedgedriver" # path to the edgedriver executable
    os.environ["webdriver.edge.driver"] = edgedriver
    
    driver = webdriver.Edge(edgedriver)
    driver.get(url)
    
    page = driver.find_element_by_xpath('//*[@id="main"]/div[8]/div/div/div[3]/div/div[1]')
    soup = BeautifulSoup(page, 'html5')
    
    driver.close()
    driver.quit()
    
    return soup    
```
