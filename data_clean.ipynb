```python
def format_price(x):
    if x[-3:] == 'Th.':
        return 1000 * int(float(x[1:-3]))
    elif x[-1:] == 'm':
        return 10**6 * int(float((x[1:-1])))
    elif x == '-':
        return -1
    else:
        int(float((x[1:])))
```

```python
# fix position
def fix_pos(x):
    if x == None:
        return 'No position'
    if x == 'Centre-Forward':
        return 'Forward'
    if x == 'Centre-Back':
        return 'Defender'
    if x == 'Central Midfield':
        return "Midfielder"
    if x == 'Defensive Midfield':
        return "Midfielder"
    if x == 'Attacking Midfield':
        return "Midfielder"
    if x =='Left-Back':
        return "Defender"
    if x =='Right-Back':
        return "Defender"
    if x =='Right Winger':
        return 'Defender'
    if x == 'Left Winger':
        return 'Forward'
    if x =='Left Midfield':
        return 'Midfield'
    if x =='Right Midfield':
        return 'Midfield'
    if x == 'Goalkeeper':
        return 'Goalkeeper'
    if x == 'Second Striker':
        return 'Forward'
    if x == 'Sweeper':
        return 'Defender'
    else:
        return x
```

```python
df['fixed_pos'] = df['position'].apply(lambda x: fix_pos(x))
```

```python
def format_age(x):
    if type(x) == str:
        return 2020 - int(''.join(x for x in x if x.isdigit()))
    if str == None:
        return 0
    return 2020 - x
```
