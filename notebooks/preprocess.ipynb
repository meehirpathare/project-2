```python
import pandas as pd
import os
import sklearn
import datetime as dt
```

```python
df = pd.read_csv('combined_data.csv')
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df = df.drop(columns = ['id_x', 'player_api_id', 'player_fifa_api_id_x', 'id_y', 'player_fifa_api_id_y', 'birth_year', 'month', 'transfer_year', 'transfer_age', 'transfer_year_low'])
```

```python
df.columns
```

```python
df['date'] = pd.to_datetime(df['date'])
df['birthday'] = pd.to_datetime(df['birthday'])
```

```python
df.head()
```

```python
foot_dummies = pd.get_dummies(df['preferred_foot'], drop_first = True)
df = pd.concat([df, foot_dummies], axis = 1)
df.drop(columns = ['preferred_foot'], inplace=True)
```

```python
attack_dummies = pd.get_dummies(df['attacking_work_rate'], drop_first = True)
attack_dummies['medium'] = attack_dummies['medium'] + attack_dummies['norm']
attack_dummies.drop(columns = ['le', 'norm', 'y'], inplace=True)
df = pd.concat([df, attack_dummies], axis = 1)
```

```python
def_dummies = pd.get_dummies(df['defensive_work_rate'], drop_first=True)
def_dummies.columns
```

```python
def_dummies['def_low'] = def_dummies['low']    
def_dummies['def_med'] = def_dummies['medium'] 
def_dummies['def_high'] = def_dummies['high'] 

lst = list(def_dummies.columns)
def_dummies.drop(columns = lst, inplace=True)
#def_dummies.drop(columns= ['2', '3', '7', '8', '_0', 'ean', 'es', 'high', 'low', 'medium', 'o', 'ormal'], inplace=True)
df = pd.concat([df, def_dummies], axis = 1)
```

```python
pos_dummies = pd.get_dummies(df['position'], drop_first = True)
pos_dummies['Midfield'] = pos_dummies['Midfield'] + pos_dummies['Midfielder'] 

pos_dummies.drop(columns = ['Midfielder'], inplace=True)
df = pd.concat([df, pos_dummies], axis = 1)
```

```python
df.drop(columns=['attacking_work_rate', 'defensive_work_rate', 'joined_team', 'team_left'], inplace=True)
```

```python
join_dummies = pd.get_dummies(df['joined_league'], drop_first=True)
df = pd.concat([df, join_dummies], axis = 1)
```

```python
df = df.drop(columns = ['stoc'])
```

```python
df.to_csv('process_df.csv')
```
