```python
def file_list(first_year, last_year):
    list_of_files = []
    for i in range(first_year, last_year + 1):
        list_of_files.append("uefa_club_coefficients/{}_uefa_club_coeff.txt".format(i))
    return list_of_files

def get_headers(soup):    
    headers = ['pos']
    for season in soup.find_all("th", class_="table_member-season"):
        headers.append(season.text)
    headers.append('Pts')
    headers.insert(1, 'name')
    return headers

def get_pos(soup):    
    pos = []
    for span in soup.find_all("td", class_ = "table_member-pos"):
        pos.append(float(span.text.strip()))
    return pos

def get_names(soup):    
    names = []
    for name in soup.find_all("td", class_ = "table_member-name"):
        names.append(name.text[100:1000].strip())
    return names

def get_scores(soup):    
    scores = []
    for item in soup.find_all("td", class_="table_member-season"):
        scores.append(item.text) 
    
    scores = np.array(scores)
    scores = np.where(scores=='-', '0', scores)
    scores = [float(x) for x in scores]
    scores = list(np.array_split(scores, len(scores)/5))
    scores =  [list(x) for x in scores]
    
    return scores

def get_tots(soup):    
    tot_scores = []
    for item in soup.find_all("td", class_="table_member-points"):
        tot_scores.append(float(item.text.strip()))
    return tot_scores
```

```python
def get_elements(file_name):
    soup = BeautifulSoup(open(file_name), 'lxml')
    
    headers = get_headers(soup)
    pos = get_pos(soup)
    names = get_names(soup)
    scores = get_scores(soup)
    totals = get_tots(soup)
    
    return [headers, pos, names, scores, totals]
```

```python
years = file_list(1997, 2020)
headers, pos, names, scores, totals = get_elements(years[0])
```
