```python
import sqlite3
import pandas as pd
import os
```

```python
con = sqlite3.connect('database.sqlite')
```

```python
df_player = pd.read_sql_query("SELECT * FROM Player", con)
```

```python
df_attr = pd.read_sql_query("SELECT * FROM Player_Attributes", con)
```

```python
df_attr.to_csv('debug.csv')
```

```python
df_player
```

```python
df_attr
```

```python
df_player = df_player.merge(df_attr,how="inner",on="player_api_id")
```

```python
df_combined = df_player
```

```python
df_combined['month'] = pd.DatetimeIndex(df_combined['date']).month
df_combined['transfer_year'] = pd.DatetimeIndex(df_combined['date']).year
```

```python
filtered_df = pd.read_csv('fixed_year.csv')
```

```python
filtered_df['transfer_year_low'] = filtered_df['transfer_year'] - 1
```

```python
filtered_df
```

```python
df_combined = df_combined.merge(filtered_df, on=['player_name', 'transfer_year'])
```

```python
df_2 = df_combined
df_2.drop(columns = ['id_x', 'player_api_id', 'player_fifa_api_id_x', 'id_y', 'player_fifa_api_id_y', 'birth_year', 'month', 'transfer_year', 'transfer_age', 'transfer_year_low'])
```
